```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import cv2
from keras.models import load_model
import numpy as np
```

```python
facedetect = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

cap=cv2.VideoCapture(0)
cap.set(3, 640)
cap.set(4, 480)
font=cv2.FONT_HERSHEY_COMPLEX
```

```python
model = load_model('khuonmat.h5')
```

```python
def get_className(classNo):
    if classNo==1:
        return "Dat_2021050821"
    elif classNo==2:
        return "Thuy_2021050673"
    elif classNo==3:
        return "Trang_2021050900"
    elif classNo==4:
        return "Thang_2021050757"
    elif classNo==5:
        return "Vuong_2021050744"
```

```python
while True:
    sucess, imgOrignal = cap.read()
    faces = facedetect.detectMultiScale(imgOrignal,1.3,5)
    for x,y,w,h in faces:
        gray_img = cv2.cvtColor(imgOrignal, cv2.COLOR_BGR2GRAY)
        crop_img = gray_img[y:y+h,x:x+w]
        img = cv2.resize(crop_img, (100,100))
        img=img.reshape(100, 100, 1)
        img = np.array(img)
        prediction=model.predict(np.array([img]))
        classIndex= np.argmax(prediction)
        probabilityValue=np.amax(prediction)
        if classIndex==1:
            cv2.rectangle(imgOrignal,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.rectangle(imgOrignal, (x,y-40),(x+w, y), (0,255,0),-2)
            cv2.putText(imgOrignal, str(get_className(classIndex)),(x,y-10), font, 0.75, (255,255,255),1, cv2.LINE_AA)
        elif classIndex==2:
            cv2.rectangle(imgOrignal,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.rectangle(imgOrignal, (x,y-40),(x+w, y), (0,255,0),-2)
            cv2.putText(imgOrignal, str(get_className(classIndex)),(x,y-10), font, 0.75, (255,255,255),1, cv2.LINE_AA)
        elif classIndex==3:
            cv2.rectangle(imgOrignal,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.rectangle(imgOrignal, (x,y-40),(x+w, y), (0,255,0),-2)
            cv2.putText(imgOrignal, str(get_className(classIndex)),(x,y-10), font, 0.75, (255,255,255),1, cv2.LINE_AA)
        elif classIndex==4:
            cv2.rectangle(imgOrignal,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.rectangle(imgOrignal, (x,y-40),(x+w, y), (0,255,0),-2)
            cv2.putText(imgOrignal, str(get_className(classIndex)),(x,y-10), font, 0.75, (255,255,255),1, cv2.LINE_AA)
        elif classIndex==5:
            cv2.rectangle(imgOrignal,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.rectangle(imgOrignal, (x,y-40),(x+w, y), (0,255,0),-2)
            cv2.putText(imgOrignal, str(get_className(classIndex)),(x,y-10), font, 0.75, (255,255,255),1, cv2.LINE_AA)

        cv2.putText(imgOrignal,str(round(probabilityValue*100, 2))+"%" ,(180, 75), font, 0.75, (255,0,0),2, cv2.LINE_AA)
    cv2.imshow("Result",imgOrignal)
    k=cv2.waitKey(1)
    if k==ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
